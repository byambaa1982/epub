# Удахгүй болох зүйлсийн талаарх эргэцүүлэл?

Хүний нийгэм бол хоорондоо яв цав нэгдэх боломжгүй хувь хүмүүсээс тогтоно. Харин машинууд бол хувь хүмүүс биш тэд амархан нэгдэж нэг цогц сүлжээ (network) үүсгэж байдаг. Одоо бид юутай тулгараад байна вэ гэвэл сая сая хүмүүс сая сая машинуудад ажлаа алдах гээд байгаа юм биш харин нэг цул бүхнийг мэддэг бүхнийг чаддаг систем биднийг орлох гээд байна л даа. Жишээ нь мянган эмч мянган цаг уулзаад ч бие биенийхээ чадвар туршлагыг бүрэн эзэмших боложгүй. Харин дэлхийн өнцөг булан бүрт өөр өөр анагаахын өгөгдөл дээр өөр өөр компаниудын хөгжүүлж буй AI-ууд бол хэдхэн минутын дотор интеграцид орж шилдэг анагаахын AI эмч болж байна. Удахгүй хөдөөний хамгийн бөглүү сум ч гэсэн хамгийн сүүлийн үеийн анагаахын ололтыг цаг алдалгүй хүртэх болно. Өөр нэг жишээ авахад 2 жолоочгүй машин 2 талаас уулзвар дээр ирлээ гэж бодоход энэ бол 2 өөр машинуудын тухай яриа биш нэг алгоритмын 2 өөр хэсэг. Нэг нь хаашаа ямар хурдтай явж байгааг мэдэж байгаа нэг тархины 2 өөр эрхтэн гэсэн үг. Хоорондоо мөргөлдөх боломж үлэмж бага. Өнөөдөр дэлхий даяар жилд 1.25 сая хүн авто осолоос нас бардаг. 90 хувь нь хүний анхаарал болгоомжгүй байдлаас болдог гэж байгаа. Энэ нь нэг жилд дайн гэмт хэрэг алан хядлагад өртсөн хүмүүсийн тоог нийлүүлснээс 2 дахин их. Асуулт. Та хувийн эмнэлэгийн эзэн бол AI хөгжүүлэхэд хөрөнгө оруулах уу эмч нараа сургах уу? Такси компаниуд жолооч нарт цалин өгсөөр байх уу хажууд нь алдаа гаргахгүй өвдөхгүй ажил таслахгүй хамаагүй мундаг шийдэл байхад?
