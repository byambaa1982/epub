# Удахгүй болох зүйлсийн талаарх эргэцүүлэл?

Хүний нийгэм бол хоорондоо яв цав нэгдэх боломжгүй хувь хүмүүсээс тогтоно. Харин машинууд бол хувь хүмүүс биш тэд амархан нэгдэж нэг цогц сүлжээ (network) үүсгэж байдаг. Одоо бид юутай тулгараад байна вэ гэвэл сая сая хүмүүс сая сая машинуудад ажлаа алдах гээд байгаа юм биш харин нэг цул бүхнийг мэддэг бүхнийг чаддаг систем биднийг орлох гээд байна л даа. 

Жишээ нь мянган эмч мянган цаг уулзаад ч бие биенийхээ чадвар туршлагыг бүрэн эзэмших боложгүй. Харин дэлхийн өнцөг булан бүрт өөр өөр анагаахын өгөгдөл дээр өөр өөр компаниудын хөгжүүлж буй AI-ууд бол хэдхэн минутын дотор интеграцид орж шилдэг анагаахын AI эмч болж байна. Удахгүй хөдөөний хамгийн бөглүү сум ч гэсэн хамгийн сүүлийн үеийн анагаахын ололтыг цаг алдалгүй хүртэх болно.

Өөр нэг жишээ авахад 2 жолоочгүй машин 2 талаас уулзвар дээр ирлээ гэж бодоход энэ бол 2 өөр машинуудын тухай яриа биш нэг алгоритмын 2 өөр хэсэг. Нэг нь хаашаа ямар хурдтай явж байгааг мэдэж байгаа нэг тархины 2 өөр эрхтэн гэсэн үг. Хоорондоо мөргөлдөх боломж үлэмж бага. Өнөөдөр дэлхий даяар жилд 1.25 сая хүн авто осолоос нас бардаг. 90 хувь нь хүний анхаарал болгоомжгүй байдлаас болдог гэж байгаа. Энэ нь нэг жилд дайн гэмт хэрэг алан хядлагад өртсөн хүмүүсийн тоог нийлүүлснээс 2 дахин их. Асуулт. Та хувийн эмнэлэгийн эзэн бол AI хөгжүүлэхэд хөрөнгө оруулах уу эмч нараа сургах уу? Такси компаниуд жолооч нарт цалин өгсөөр байх уу хажууд нь алдаа гаргахгүй өвдөхгүй ажил таслахгүй хамаагүй мундаг шийдэл байхад?








AI-ийн хувьсгал эрч хүчээ авсаар байгаа тул дараагийн хагас арван жилд хиймэл оюун ухаанд үзүүлэх өөрчлөлтийн нөлөөг төсөөлж буй зарим ухаалаг статистикийг эндээс үзнэ үү.

- Эдийн засгийн өсөлт: PwC-ийн тайланд дурдсанаар хиймэл оюун ухааны технологи нь 2030 он гэхэд дэлхийн эдийн засагт 15.7 их наяд ам.долларын хувь нэмэр оруулж, дэлхийн ДНБ-ий 14 гаруй хувиар өснө.

- Ажлын байрны өөрчлөлт: Хиймэл оюун ухаан нь ажлын байрыг орлох талаар санаа зовж байгаа ч шинэ ажлын байр бий болгох төлөвтэй байна. Гартнер 2025 он гэхэд хиймэл оюун ухаан нь 2 сая шинэ ажлын байр бий болгоно гэж таамаглаж байгаа нь хөдөлмөрийн зах зээлийн бууралтаас илүү хувьслыг онцолж байна.

- Эрүүл мэндийн үйлчилгээ: AI нь зөвхөн АНУ-д л гэхэд эрүүл мэндийн зардлыг жил бүр 100 тэрбум доллараар бууруулахад тусална гэж McKinsey тооцоолсон. Энэ нь хувь хүний ​​​​эмчилгээ, эрт оношлох, үр дүнтэй эмийн нээлтийг сайжруулснаар ирдэг.

- Эрчим хүч хэмнэлт: Capgemini-ийн судалгаагаар хиймэл оюун ухаан нь ойрын 5 жилд үйлдвэрлэл зэрэг салбарт эрчим хүчний хэрэглээг 20% хүртэл бууруулж чадна гэдгийг харуулж байна.

- Жижиглэнгийн худалдааны хувьсгал: IBM-ийн мэдээлснээр 2025 он гэхэд харилцагчийн харилцааны 85% нь хүнгүйгээр зохицуулагдана. Энэ нь чатбот гэх мэт хиймэл оюун ухаанд суурилсан харилцагчийн үйлчилгээний хэрэгсэл рүү шилжсэн гэсэн үг.

- Автомат жолоодлого: Статиста 2025 он гэхэд бие даасан тээврийн хэрэгслийн дэлхийн зах зээлийн хэмжээ 556 тэрбум ам.долларт хүрч, тээврийг үндсээр нь өөрчлөх болно гэж Статиста төлөвлөж байна.

- Хөдөө аж ахуй ба хүнс: ХАА-н хиймэл оюун ухаан 2020-2025 он хүртэл CAGR 25%-иас дээш өсөх төлөвтэй байгаа бөгөөд хиймэл оюун ухаанд суурилсан шийдлүүд нь ургацын ургацыг урьдчилан таамаглах, хортон шавьж илрүүлэх, тогтвортой газар тариалан эрхлэхэд тусалдаг.

- Суралцах ба боловсрол: Цахим сургалтын платформууд нь хувийн сургалтын туршлагыг бий болгох зорилгоор хиймэл оюун ухааныг нэгтгэж байна. HolonIQ тайланд 2025 он гэхэд дижитал сургалтын платформуудын 90% нь хиймэл оюун ухааныг ашиглан оюутнуудад тохирсон контентыг хангана гэж мэдэгджээ.

Дүгнэж хэлэхэд:

Хиймэл оюун ухаан бидний амьдралын бүх талбарт илүү гүнзгий нэвтэрч байгаа тул ирэх 5 жил өөрчлөгдөнө гэж амлаж байна. Асуудал нь нийгмийн бүхий л салбарт үр өгөөжийг нь тэгш хуваарилахын тулд түүний чадавхийг хариуцлагатай ашиглах явдал байх болно.
