# Дата дата л гэнэ яг юу юм бэ?

Америкийн хамгийн том сүлжээ дэлгүүрүүдийн нэг Таргетад нэгэн аав их л ууртай орж ирээд танай менежиртэй уулзмаар байна гэж. Менежир ирмэгц тэрээр та нар миний арван жилийн охин руу хүүхдийн живх өлгийний сурталчилгааны хуудас явуулдаг нь яаж байгаа юм жаахан амитаныг төр гээд байгаа юм уу гээд л агсарч гарч. Менежир манай системд алдаа гарсан байх дахин ийм алдаа гарахгүй гээд зөндөө уучлалт гуйж тайвшруулаад явуулж. Гэтэл хэд хоногийн дараа нөгөө хүн ахин хүрч ирээд манай охин жирэмсэн төрөх дөхсөн бид мэдээгүй явж харин та нар яаж мэдэв гэж. Менежир дата мэргэжилтнүүдээ дуудаж тайлбарлуулав. Тэд танай охин 7 сарын өмнө манайхаас жирэмсний тест худалдаж авчээ дараа нь манайхаас дахин ариун цэврийн хэрэглэл аваагүй байна. Энэ нь манай системийн хувьд танай охинийг жирэмсэн гэж таних өндөр магадлал болжээ гэж. Таргет бол таныг юу хүсэж байгааг та өөрөө ч мэдэхээс өмнө мэддэж байдаг. Та олон жил үйлчлүүлж байгаа бол зун дөхөх үед таны өвс хадагч хуучирсан одоо шинийг авах цаг болсон гэж та өөрөө санахаас өмнө яг өвс хадаж эхлэх үед тань руу сүпер хямдралынхаа купоныг явуулна. Тэр бүү хэл та эхнэрээсээ салсан, байшингаа томруулсан, эсвэл жижигрүүлсэн, албан тушаал ахисан буусан бүгдийг мэдэж байгаа. Яаж? Дата. Жишээ нь та хүүхдийн өглөөний хоол хувцас авдаг байснаа болиод дараалан хэдэн өдөр пиво бэлгэвч аваад эхлэвэл тэдний систем таныг эхнэрээсээ салсан гэж үзээд тань руу ахин хүүхдийн хувцасны сурталчилгаа явуулахгүй оронд нь насанд хүрэгчидийн бүтээгдэхүүнүүд санал болгох болов уу. Хэрэв та машинаас яг нэг жилийн дараа би хаана юу хийж байх вэ гэж асуувал мэдэхгүй хэлэх боломжгүй харин яг тань шиг ажил орлоготой мянга мянган хүмүүс яг жилийн дараа хэдэн хувь нь хаана юу хийж байхыг бол 99% хувь нарийчлалтай таамаглаж чадна. Энэ бол ид шид биш энэ бол датаны хүч чадал. 

Хэдхэн жилийн дараа дата олж авсан тэрийг боловсруулж ашигласан компани нь чадаагүйгээ зах зээл дээрээс арчиж хаях болно. Тавилан нь тийм юм. Одоо бол газрын тосны төлөө биш датаны төлөө уралдаан эхэлсэн. Тиймээс компани болгон дата инженертэй болох ёстой юм. Үнэтэй гэж айлгүй клауд платформуудад хөрөнгө оруулах ёстой. Тун удах жижигхээн гарааны бизнесүүд Номин, Таван богдыг зах зээлээс шахаж мэдэх юм. Хэрэв тэд датаны хүч чадалыг дутуу үнэлсээр л байвал шүү дээ. 

Өнгөрөсөн сарын 24-30 хооронд Сан-Франциско хотод `Data+AI Summit` болж онлайн оффлайн 70 гарай мянган дата мэргэжилтнүүд оролцлоо. Amazon, Microsoft, JP Morgan гээд аварга компаниуд дата хиймэл оюунд ямар их ач холбогдол өгч байна. Технологийн хөгжил ч толгой эргэм хурдтай байна. Саяхан л од болж байсан data lake data warehouse технологиуд lakehouse т байраа тавьж өгч байна. Зөвхөн англиар бичихэд өгүүлбэрүүдийг python SQL код болгож хувиргаад дата боловсруулж график дүрслэл хийчихэж байна. Боловсруулж байгаа датаны хэмжээ дамжаа ч гэж тоймгүй юм болж. JP Морган 500 PB датаг өдөрт боловсруулж ашигладаг. Хэрэв энэ датаг CD хуулаад өрөх юм бол Эвэристийн оргилоос 100 дахин өндөр болно гэж захирал нь саммит дээр хэлсэн. Databricks зэрэг клауд технологи ашиглахгүйгээр ийм хэмжээний датаг зохицуулна гэдэг ямар ч боломжгүй зүйл юм. Энэ зуун бол датаизмийн зуун юм.
