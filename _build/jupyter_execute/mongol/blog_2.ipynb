# Дата дата л гэнэ яг юу юм бэ?

Америкийн хамгийн том сүлжээ дэлгүүрүүдийн нэг Таргетад нэгэн аав их л ууртай орж ирээд танай менежиртэй уулзмаар байна гэж. Менежир иртэл тэрээр та нар миний арван жилийн охин нэрээр манайх руу хүүхдийн живх өлгийний сурталчилгааны хуудас явуулдаг нь яаж байгаа юм төр гээд байгаа юм уу гээд л агсарч гарч. Менежир манай системд алдаа гарсан байх дахин ийм алдаа гарахгүй гээд зөндөө уучлалт гуйж тайвшруулаад явуулж. Гэтэл хэд хоногийн дараа нөгөө хүн ахин хүрч ирээд манай охин жирэмсэн төрөх дөхсөн бид мэдээгүй явж харин та нар яаж мэдэв гэж. Менежир дата мэргэжилтнүүдээ дуудаж тайлбарлуулав. Тэд танай охин 7 сарын өмнө манайхаас жирэмсний тест худалдаж авчээ дараа нь манайхаас дахин ариун цэвэрийн хэрэглэл аваагүй байна. Энэ манай системийн хувьд танай охинийг жирэмсэн гэж үзэх өндөр магадлал болжээ гэж. Таргет бол таныг юу хүсэж байгаагаа мэдэхээс өмнө мэддэж байдаг. Та олон жил үйлчлүүлж байгаа бол зун дөхөх үед таны өвс хадагч хуучирсан одоо шинийг авах цаг болсон гэж мэдэж байгаа гэсэн үг. Тэр бүү хэл та эхнэрээсээ салсан, байшингаа томруулсан, эсвэл жижигрүүлсэн, албан тушаал ахисан буусан бүгдийг мэдэж байгаа. Яаж? Дата. Жишээ нь та хүүхдийн өглөөний хоол хувцас авдаг байснаа болиод пиво бэлгэвч аваад эхлэвэл тэдний систем таныг эхнэрээсээ салсан гэж үзээд тань руу ахин хүүхдийн хувцасны сурталчилгаа явуулахгүй оронд нь насанд хүрэгчидийн бүтээгдэхүүнүүд санал болгох болов уу. 

Хэдхэн жилийн дараа дата олж авсан тэрийг боловсруулж ашигласан компани нь чадаагүйгээ зах зээл дээрээс арчиж хаях болно. Тавилан нь тийм юм. Одоо бол газрын тосны төлөө биш датаны төлөө уралдаан эхэлсэн. Тиймээс компани болгон дата инженертэй болох ёстой юм. Үнэтэй гэж айлгүй клауд платформуудад хөрөнгө оруулах ёстой. Тун удах жижигхээн гарааны бизнесүүд Номин, Таван богдыг зах зээлээс шахаж мэдэх юм. Хэрэв тэд датаны хүч чадалыг дутуу үнэлсээр л байвал шүү дээ.
